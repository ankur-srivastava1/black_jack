```python
import random
```

```python
#global_variable
all_suits = ['Clubs','Spades','Diamonds','Hearts']
all_faces = ['Ace',2,3,4,5,6,7,8,9,10,'Jack','Queen','King']
deck = [(x,y) for x in all_suits for y in all_faces]
```

```python
class Card:
    
    def __init__(self):
    
        global deck
        randomCard = random.choice(deck)
        deck = list(set(deck) - set(randomCard))
        
        self.suit = randomCard[0]
        self.face = randomCard[1]
        
    def value(self, playerTotal = 0):
        face = self.face
        if (face == 'Jack' or face == 'Queen' or face == 'King'):
            val = 10
        elif (face == 'Ace'):
                val = 1
        else:
            val = face
        
        return(val)
    
    def show(self):
        if self.suit == 'Hearts':
            print(f"{self.face} \u2665")
        elif self.suit == 'Diamonds':
            print(f"{self.face} \u2666")
        elif self.suit == 'Spades':
            print(f"{self.face} \u2660")
        else:
            print(f"{self.face} \u2663")
    
```

```python
class Hand(Card):
    
    def __init__(self,player_name):
        Card.__init__(self)
        self.player_name = player_name
        
    def draw(self):
        hand = [Card(),Card()]
        return(hand)
    
    def ace_present(self,hand):
        ace_present_flag = False
        for i in range(len(hand)):
            if hand[i].face == 'Ace':
                ace_present_flag = True
                break
        return(ace_present_flag)
        
    
    def get_total_value(self,hand, dealerBetting = False):
        total_value = 0
        ace_present_flag = False
        if (self.player_name == 'Dealer' and dealerBetting == False):
            for i in range(len(hand)-1):
                total_value = total_value + hand[i].value(total_value)
                if hand[i].face == 'Ace':
                    ace_present_flag = True
            if ace_present_flag == True and total_value + 10 < 21:
                total_value = total_value + 10
                
        elif (self.player_name == 'Dealer' and dealerBetting == True):
            for i in range(len(hand)):
                total_value = total_value + hand[i].value(total_value)
                if hand[i].face == 'Ace':
                    ace_present_flag = True
            if ace_present_flag == True and total_value + 10 < 21:
                total_value = total_value + 10
        else:
            for i in range(len(hand)):
                total_value = total_value + hand[i].value(total_value)
                if hand[i].face == 'Ace':
                    ace_present_flag = True
            if ace_present_flag == True and total_value + 10 < 21:
                total_value = total_value + 10

        return(total_value)
    
    def show_hand(self,hand,toal_value, dealerBetting = False):
    
        if (self.player_name == 'Dealer' and dealerBetting == False):
            print("Dealer Cards: ")
            for n in range(len(hand) - 1):
                hand[n].show()
            print("Face Down Card")
            print("Dealer Total: {}".format(toal_value))
        elif (self.player_name == 'Dealer' and dealerBetting == True):
            print("Dealer Cards: ")
            for n in range(len(hand)):
                hand[n].show()
            print("Dealer Total: {}".format(toal_value))
        else:   
            print("{} Cards: ".format(self.player_name))
            for n in range(len(hand)):
                hand[n].show()
            print("Total: {}".format(toal_value))

    def call_it(self):
        valid_input = False
        while valid_input == False:
            call = input(f"{self.player_name} Call it: Hit me or Stay: " )
            call = call.upper()
            if (call == 'HIT ME' or call == 'STAY'):
                valid_input = True
        return(call)
    
    def replay_game(self):
        print('\n')
        tmp = input('{} Enter Y to replay or N to quit: '.format(self.player_name))
        if tmp.upper() == 'Y':
            return(True)
        else:
            return(False)
```
